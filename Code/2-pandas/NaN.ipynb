# NaN - Not a Number

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
n = np.nan
type(n)
```

```python
# 任何数值（整数、浮点数、0）与 NaN 作运算结果都是 NaN
m = 1
m + n
```

## NaN in Series

```python
s1 = pd.Series([1, 2, np.nan, 3, 4], index=['A', 'B', 'C', 'D', 'E'])
s1
```

```python
# 判断 Series 中的值是否是 NaN
s1.isnull()
```

```python
# 判断 Series 中的值是否不是 NaN
s1.notnull()
```

```python
 # 删除 Series 中的值是 NaN 的数据
s1.dropna()
```

## Nan in DataFrame

```python
df1 = pd.DataFrame([[1, 2, 3], [np.nan, 5, 6], [7, np.nan, 9], [np.nan, np.nan, np.nan]])
df1
```

```python
df1.isnull()
```

```python
df1.notnull()
```

```python
df1.dropna(axis=0)
```

```python
df1.dropna(axis=1)
```

```python
df1.dropna(how='all')
```

```python
df2 = pd.DataFrame([[np.nan, 5, 6], [7, np.nan, np.nan], [np.nan, np.nan, np.nan]])
df2
```

```python
# thresh 表示范围 删除少于 3 个非 NaN 值的行或列
df2.dropna(thresh=2)
```

```python
# 替换值为 NaN 的值
df2.fillna(value=1)
```

```python
# 按列替换值为 NaN 的值
df2.fillna(value={0:0, 1:1, 2:2})
```
